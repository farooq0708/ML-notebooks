```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
import math
from sklearn.metrics import r2_score, mean_sq
```

```python
median_bedrooms = math.floor(df.bedrooms.mean())
median_bedrooms



df['bedrooms'] = df['bedrooms'].fillna(median_bedrooms)
df

X = df[['area','bedrooms','age']]
y = df['price']


model = linear_model.LinearRegression()
model.fit(X,y)

model.coef_


print(model.intercept_)

y_pred = model.predict(X) 


r2 = r2_score(y, y_pred)
if r2 > .8 :
    print("Model is generalized !,")
mse = mean_squared_error(y, y_pred)

# Root Mean Squared Error
rmse = np.sqrt(mse)

print("RÂ² Score:", r2)
print("MSE:", mse)
print("RMSE:", rmse)


# plot 
plt.scatter(df['area'], df['price'], color='red', marker='o')
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Area vs Price')
plt.show()

# Bedrooms vs Price
plt.scatter(df['bedrooms'], df['price'], color='blue', marker='o')
plt.xlabel('Bedrooms')
plt.ylabel('Price')
plt.title('Bedrooms vs Price')
plt.show()

# Age vs Price
plt.scatter(df['age'], df['price'], color='green', marker='o')
plt.xlabel('Age')
plt.ylabel('Price')
plt.title('Age vs Price')
plt.show()
```
