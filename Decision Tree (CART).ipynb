```python
import numpy as np 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,classification_report
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r'C:/Users/User/Desktop/ML/ML_datasets/heart_failure_clinical_records_dataset.csv')
data.head()
```

```python
data.info
```

```python
data.isnull().sum()
data.columns.str.strip()
```

```python
X = data.iloc[:,:-1]  #(all rows and columns except last one which  will be assigned to output )
y = data.iloc[:,-1]   #All rows of last columns so only last colum as output y .

#print(X,y)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.25,random_state = 42)
```

```python
Model = DecisionTreeClassifier(max_depth=3, min_samples_split=5, min_samples_leaf=5)
Model
```

```python
Model = Model.fit(X_train,y_train)

y_pred = Model.predict(X_test)

accuracy = accuracy_score(y_test,y_pred)

accuracy
```

```python
plt.figure(figsize=(18,10))
tree.plot_tree(Model, feature_names=X.columns)
plt.show()
```
