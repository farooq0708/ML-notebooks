```python
import requests
import pandas as pd

api_key = "ee20f89faacb1f69ad838a0f2eff78b1a"

page = 1
rows = []

while True:
    url = f"https://api.themoviedb.org/3/discover/movie?api_key={api_key}&page={page}"
    response = requests.get(url)

    if response.status_code != 200:
        print("Request failed at page", page)
        break

    data = response.json()

    #  STOP when there are no more results
    if len(data["results"]) == 0:
        print("Finished. No more pages.")
        break
    

    df = pd.DataFrame(data["results"])[
        ["id", "vote_count", "popularity", "overview"]
    ]

    rows.extend(df.to_dict("records"))

    print("Fetched page:", page)
    page += 1   # THIS is what moves to the next page

df = pd.DataFrame(rows)
df.head()
```
