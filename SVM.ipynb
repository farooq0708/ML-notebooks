```python
import numpy as np 
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn import metrics
import matplotlib.pyplot as plt 

```

```python
df = pd.read_csv("C:/Users/User/Desktop/ML/ML_datasets/Social_Network_Ads.csv",encoding = "latin1")
df.head(5)
```

```python
df.shape
```

```python
X = df.iloc[:,[2,3]]
y = df.iloc[:,4]
#print(X.head)
print(y.head(5))
```

```python
#splitting data into training and testing set 
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size = .25,random_state = 0 )
```

```python
print(X_train.shape)
print(y_test.shape)
```

```python
print(y_train)
```

```python
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.fit_transform(X_test)
```

```python
classifier = SVC(kernel = 'linear',random_state = 0 )
classifier.fit(X_train,y_train)
y_pred = classifier.predict(X_test)
y_pred

```

# ('Accuracy Score : With Linear Kernel')

```python
print('Accuracy Score : With Linear Kernel')
print(metrics.accuracy_score(y_test,y_pred))
```

```python
classifier = SVC(kernel = 'rbf',random_state = 0 )
classifier.fit(X_train,y_train)
y_pred = classifier.predict(X_test)
y_pred
```

# ('Accuracy Score : With rbf Kernel')

```python
print('Accuracy Score : With rbf Kernel')
print(metrics.accuracy_score(y_test,y_pred))
```

```python
classifier = SVC(kernel = 'rbf',gamma = 15,C = 7,random_state = 0 )
classifier.fit(X_train,y_train)
y_pred = classifier.predict(X_test)
y_pred
```

# ('Accuracy Score : With Default rbf Kernel')

# ('Accuracy Score : With polynomial Kernel')

# Hyperplane Equation

```python
# -----------------------------------------------------------
# SVM Hyperplane Explanation
# -----------------------------------------------------------
# General Equation of Hyperplane:
#     w^T x + b = 0
#
# where:
#     w = [w1, w2, w3, ..., wn]  -> weight vector
#     x = [x1, x2, x3, ..., xn]  -> input feature vector
#     b -> bias/intercept
#
# Expanded Form:
#     w1*x1 + w2*x2 + w3*x3 + ... + wn*xn + b = 0
#
# -----------------------------------------------------------
# For Two Features (2D Case):
#     w1*x1 + w2*x2 + b = 0
#
# Shift all terms except x2 to the right side:
#     w2*x2 = -w1*x1 - b
#
# Divide both sides by w2 to isolate x2:
#     x2 = -(w1/w2)*x1 - (b/w2)
#
# -----------------------------------------------------------
# Identifications:
#     m = -w1 / w2       -> slope of the SVM decision boundary
#     c = -b  / w2       -> intercept of the SVM decision boundary
#
# Therefore, the final SVM boundary line equation is:
#     x2 = m*x1 + c
#
# -----------------------------------------------------------
# Comparison with the Standard Line Equation:
#     Standard line form:  y = m*x + c
#     SVM boundary form:   x2 = m*x1 + c
#
# Both equations share the same structure.
# In SVM, 'm' and 'c' are computed from the model's learned
# parameters (w1, w2, and b) rather than being chosen manually.
#
# -----------------------------------------------------------
# Decision Rule:
#     if (w^T x + b) > 0  -> Class 1
#     if (w^T x + b) < 0  -> Class 0
#     if (w^T x + b) = 0  -> Point lies exactly on the boundary line
# -----------------------------------------------------------
```
